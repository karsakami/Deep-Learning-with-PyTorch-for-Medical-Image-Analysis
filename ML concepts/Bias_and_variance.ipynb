##It's not that important that our **squiggly line** did perfect job on training dataset when it performed teribbly on test set.
###Difference between fitting between datasets its called **variance** 

So - **linear regression** has high bias (cant fit data precise) and low variance, and **squiggly line** has oposite parameters.

So Squiggly Line is **overfitting** because it adapts to the training data too much, and then perform usually bad on testing data

Ideal model has **low bias** and **low variance**, we have to find sweetspot, for this we need to use methods like **regularization**, **boosting**, **bagging**
