# **Overfitting** and **underfitting** - *two scenarios of bugs that we can experience*

######https://towardsdatascience.com/overfitting-and-underfitting-principles-ea8964d9c45c

### These terms are related to **bias** and **variance**
---
##**Overfitting** - model fits **too much** to the noise from the data.
*(Kiedy model za bardzo przyzwyczaja sie do danych treningowych)* \
Results can be noticed in **low error on training sets** but **high error on test/validation sets** - so low bias, but high variance

Overfitted model graph looks like as influenced by the wind. \
If we take a new point, our model is doing well on trening dataset, but bad on test dataset

#**Good model**
Good model will try to fit general trend of dataset, like average values, its like more smooth, more stable, without crazy variance

## **Underfitting**
Our model is doing bad in respect to training data and test data, we defenitely dont want this.

Underfitting appear when we just simplify the problem, we just use linear regression, but the problem is more complex. Lets maybe use a CNN solutions.

**Underfitting features:**

*   Models didnt learned well from the training data and performed poorly on test data, didnt get a trend
*   low variance, high bias

# We can see similarity in learning by a normal student, lets understand this intuitively.

We have student whos preparing for exam, and he got syllabus (which topics are covered on exam), if our student prepared good on exam and he got good grade at exam, thats the student whos good = the model that we want.


But he also can be confident about his preparation but perform bad at exam, thats **overfitting**, he prepared for training data (syllabus) too much. He didnt catch the trend, what it's all about.

And last case is when he performed bad at training data, and also ofcourse bad at test data - **underfitting**

**What is data leakage?**

When exams just leaked, when data is avaible for student, but this is not good because he will fail next exams

How we can see underfitting and overfitting when we deal with multi dimenstional data sets?

* Good model has the less error, the more is training time (epochs), in the context of training and test data
* Bad model is opposite, the error will increase. \
For example if model is overfitting, he perform well on training data, but error starts to increase when performed on test data. So we can capture the moment when error starts to increase and stop to train
